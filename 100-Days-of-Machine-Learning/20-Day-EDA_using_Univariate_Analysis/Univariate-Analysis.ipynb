## 1. Categorical Data

#### a. CountPlot

```python
sns.countplot(df['Survived'])
df['Survived'].value_counts().plot(kind='bar')
```

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```
