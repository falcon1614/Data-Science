### BarPlot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({"Roll":[1,2,3,4,5],"Name":["jayant","Ankit","Sachin","Shyam","Raghu"]})
```

```python
plt.bar(df['Roll'],df['Name'])
plt.show()
```

```python
plt.barh(df['Roll'],df['Name'],color='Green')
```

```python
sns.barplot(x=df['Roll'],y=df['Name'],data=df,color='pink',orient="h")
```
